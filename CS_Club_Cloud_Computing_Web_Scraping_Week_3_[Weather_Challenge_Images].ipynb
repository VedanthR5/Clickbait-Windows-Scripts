```python
# Imports

import requests
from bs4 import BeautifulSoup
```

```python
page = requests.get("https://forecast.weather.gov/MapClick.php?lat=37.7772&lon=-122.4168") # Key things to note here: lat/lon input
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
seven_day = soup.find(id="seven-day-forecast") # Using tags to locate
forecast_items = seven_day.find_all(class_="tombstone-container")
print(forecast_items)
```

```python
for forecast in forecast_items: # Getting image link
  img = forecast.find("img")
  desc = img['src']
  print(desc)
```
